```python
# Read in image
colour_img = cv.imread("image.png", 1)
plt.imshow(colour_img)
```

```python
colour_img_channels_reversed = colour_img[:, :, ::-1]       #Changing The Colour Format for open cv from BRG to RGB
plt.imshow(colour_img_channels_reversed)    
```

```python
colour_img1 =cv.imread("image.png", cv.IMREAD_COLOR)
colour_img1 = cv.cvtColor(colour_img1, cv.COLOR_BGR2RGB)
plt.imshow(colour_img1)
```

```python
cropped_region = colour_img1[0:200, 0:300]      #Cropping the image 
plt.imshow(cropped_region)
```

```python
resize_colour_img=cv.resize(colour_img1, None, fx=2, fy=2)
plt.imshow(resize_colour_img)
```

```python
desired_width = 100
desired_height = 200
dim = (desired_width, desired_height)

# Resize background image to sae size as logo image
resized_clour_img = cv.resize(resize_colour_img, dsize=dim, interpolation=cv.INTER_AREA)        #interpoling used to calculate the pixel while resizing 
plt.imshow(resized_clour_img)
```

```python
f = cv.imread('image.png')

# Example coordinates (change these!)
image = cv.circle(f, center=(100, 100), radius=15, color=(0, 0, 255), thickness=10)   # A - Blue
image = cv.circle(image, center=(200, 100), radius=15, color=(0, 255, 0), thickness=10) # B - Green
image = cv.circle(image, center=(100, 200), radius=15, color=(255, 0, 0), thickness=10) # C - Red
image = cv.circle(image, center=(200, 200), radius=15, color=(255, 255, 0), thickness=10) # D - Yellow

# Convert BGR  RGB for matplotlibx
plt.imshow(cv.cvtColor(image, cv.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```
